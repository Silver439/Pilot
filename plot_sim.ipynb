```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
import os
import numpy as np
data = "data1"
if data=="data1":
    text = "group"
else:
    text = "random"

# 定义文件夹路径和CSV文件列表
data_folder = os.path.join('Data', data)
#files = ['50_12_0_pre.csv', '50_12_2_pre.csv', '50_12_20_pre.csv', '50_12_Uni_pre.csv']
files = ['50_12_0_post.csv', '50_12_2_post.csv', '50_12_20_post.csv', '50_12_Uni_post.csv']
curve_names = ['β = 0', 'β = 2', 'β = 20', 'Uniform']

# 为每个条件预定义颜色（使用Plotly默认颜色序列）
colors = px.colors.qualitative.Plotly[:len(curve_names)] 

# 创建两个DataFrame分别存储原始均值和最优均值
all_mean_data = pd.DataFrame()
all_best_data = pd.DataFrame()

# 循环处理每个文件
for file, name in zip(files, curve_names):
    file_path = os.path.join(data_folder, file)
    df = pd.read_csv(file_path, index_col='Repetition')
    
    # 计算每个step的均值
    mean_results = df.mean(axis=0)
    
    # 计算当前最优均值（累积最小值）
    best_results = np.minimum.accumulate(mean_results.values)
    
    # 添加到原始均值DataFrame
    temp_mean_df = pd.DataFrame({
        'Iter': range(1, 201),
        'Value': mean_results.values,
        'Condition': name
    })
    
    # 添加到最优均值DataFrame
    temp_best_df = pd.DataFrame({
        'Iter': range(1, 201),
        'Value': best_results,
        'Condition': name
    })    
    all_mean_data = pd.concat([all_mean_data, temp_mean_df])
    all_best_data = pd.concat([all_best_data, temp_best_df])

# 创建子图（2行1列，垂直排列）
fig = make_subplots(
    rows=2, cols=1,
    vertical_spacing=0.1  # 调整子图间距
)

# 为每个条件添加原始均值曲线（到第一个子图）
for i, condition in enumerate(curve_names):
    mean_df = all_mean_data[all_mean_data['Condition'] == condition]
    fig.add_trace(
        go.Scatter(
            x=mean_df['Iter'],
            y=mean_df['Value'],
            name=condition,
            legendgroup=condition,  # 分组图例
            showlegend=True,
            line=dict(color=colors[i]),  # 强制指定颜色
            meta=[condition],
            hovertemplate="Iter: %{x}<br>Mean: %{y:.4f}<br>Condition: %{meta[0]}<extra></extra>"
        ),
        row=1, col=1
    )

# 为每个条件添加最优均值曲线（到第二个子图）
for i, condition in enumerate(curve_names):
    best_df = all_best_data[all_best_data['Condition'] == condition]
    fig.add_trace(
        go.Scatter(
            x=best_df['Iter'],
            y=best_df['Value'],
            name=condition,
            legendgroup=condition,  # 分组图例（与原始均值共享）
            showlegend=False,      # 避免重复显示图例
            line=dict(color=colors[i]),  # 使用相同的颜色
            meta=[condition],
            hovertemplate="Iter: %{x}<br>Best: %{y:.4f}<br>Condition: %{meta[0]}<extra></extra>"
        ),
        row=2, col=1
    )

# 更新布局
fig.update_layout(
    title_text=text+"_50_12_globalBased",            # 添加总标题
    title_x=0.5,                   # 标题居中
    height=1080,                    # 减少高度
    width=1920,                     # 增加宽度
    template='plotly_white',
    legend_title_text='Condition',
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.95)
)

# 更新Y轴和X轴标签
fig.update_yaxes(title_text="Instant Regret", row=1, col=1)
fig.update_yaxes(title_text="Simple Regret", row=2, col=1)
fig.update_xaxes(title_text="Iteration", row=2, col=1)  # 仅底部子图显示X轴标签

# 显示图形
fig.show()

# 保存为HTML文件
pio.write_html(fig, '50_12_globalBased.html')
```
